```python
import platform
import subprocess

def get_m_processor_info():
    # Use subprocess to run the `sysctl` command, which retrieves system information on macOS
    try:
        processor_info = subprocess.check_output(["sysctl", "machdep.cpu.brand_string"]).strip().decode()
        return processor_info
    except subprocess.CalledProcessError:
        return "Unable to determine processor information."

# Display the processor info
processor_info = get_m_processor_info()
print(f"Processor: {processor_info}")
```
